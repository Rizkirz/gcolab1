# Selection OP

```python
import pandas as pd
import numpy as np
```

```python
n_rows = 10
n_cols = 5
cols = tuple('ABCDE') #convert string ke tuple

df = pd.DataFrame(np.random.randint(1,5, size=(n_rows, n_cols)),
                  columns = cols)
df
```

```python
#Selection dengan operator logika OR |
df[(df['A'] == 1) | (df['A'] == 3)]
```

```python
#selection isin 
df[df['A'].isin([1,3])]
```

```python
#negasi (~) kebalikan isin 
df[~df['A'].isin([1,3])]
```

# Konversi string ke numerik 

```python
import pandas as pd
```

```python
data = {'col1': ['1', '2', '3', 'abc'],
        'col2':['1', '2', '3','4']}
df = pd.DataFrame(data)
df
```

```python
df.dtypes
```

```python
#Konversi tipe data dengan fungsi .astype()
df_x = df.astype({'col2' : 'int'})
df_x
```

```python
df_x.dtypes
```

```python
df.apply(pd.to_numeric, errors='coerce')
#coerce = ketika mengconversi ke numerik terdapat data yang tidak bisa di konversi ke numerik data tersebut menjadi NaN
```
